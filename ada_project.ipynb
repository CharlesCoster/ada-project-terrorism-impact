```
cpds = pd.read_excel('http://www.cpds-data.org/images/Update2019/CPDS-1960-2017-Update-2019.xlsx', parse_dates=True)
```

```
gov_comp = pd.read_excel('http://www.cpds-data.org/images/Update2019/Government-Composition-1960-2017-Update-2019.xlsx', parse_dates=True)
```

```
cpds.head()
```

Construct a reduce dataset with 3 countries of interest for now

```
cpds_red = cpds[cpds.country.isin(['France','Germany','United Kingdom'])]
```

Plot real GDP for these countries from 1960 to 2017



```
cpds_red.set_index('year').groupby('country')['realgdpgr'].plot(legend=True)
```

Plot the polulations for these 3 countries

```
cpds_red.set_index('year').groupby('country')['pop'].plot(legend=True)
```

```
for c in ['France', 'United Kingdom', 'Germany'] :
  df = cpds_red[cpds_red.country.isin([c])]
  df.set_index('year').groupby('country')['gov_right1','gov_cent1','gov_left1'].plot(legend=True,title=c)
plt.show()
```

```
for c in ['France', 'United Kingdom', 'Germany'] :
  df = cpds_red[cpds_red.country.isin([c])]
  df.set_index('year').groupby('country')['gov_right2','gov_cent2','gov_left2'].plot(legend=True,title=c)
plt.show()
```

```
for c in ['France', 'United Kingdom', 'Germany'] :
  df = cpds_red[cpds_red.country.isin([c])]
  df.set_index('year').groupby('country')['gov_right3','gov_cent3','gov_left3'].plot(legend=True,title=c)
plt.show()
```
